```python
# 導入必要的模組
import os
import json
import pandas as pd
import re
import pickle
from sklearn import preprocessing
```

```python

def read_json_file(file):
    with open(file, 'r') as f:
        return json.load(f)
# 定義找到特定目錄中所有JSON檔案的函數
def find_json_files(directory):
    json_files = []
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith('.json'):
                json_files.append(os.path.join(root, file))
    return json_files
def create_summary_df(directory:str):
    all_names = []
    all_summaries = []
    
    json_files = find_json_files(directory)
    for json_file in json_files:
        with open(json_file, "r", encoding="utf-8") as f:
            data = json.load(f)
        for name, summary_list in data.items():
            all_names.extend([name] * len(summary_list))
            all_summaries.extend(summary_list)

    summary_df = pd.DataFrame({"name": all_names, "summary": all_summaries})
    return summary_df
```

```python
directory = "G:\共用雲端硬碟\FBI_Intern\content" # 您的資料夾路徑
summary_df = create_summary_df(directory)
summary_df = summary_df[['name', 'summary']]
summary_df.head()
```
